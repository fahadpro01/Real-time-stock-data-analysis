```python
import yfinance as yf
msft = yf.Ticker("MSFT") #  A " tick " is any change in the price of the security, whether that movement is up or down.

print(msft.history(period="5y")) #Microsoft Corp stock


```

```python
import seaborn as sns
import matplotlib.pyplot as plt


hist = msft.history(period="5y")
hist['Close'].plot(figsize=(16,9))
```
